## Heatmap with all features ordered by tissue type 

prepare data

```r
md$TissueTreatment <- paste(md$Tissue, md$Treatment, sep = "_")
```

```r
head(md)
```

```r
nonfeatures <- c(1:9) # define which columns are metadata
tab <- cbind(md[match(ft$SampleNo,md$SampleNo),], ft)
tab[,-nonfeatures] <- t(apply(tab[,-nonfeatures], 1, function(x) (x)/sum((x)))) # normalize by TIC
tab <- tab[with(tab, order(Tissue,Treatment, TreatmentIndiv)), ]
```

```r
tab[1:5,c(1,10:15)]
```

```r
ft_plot <- tab[,c((length(nonfeatures)+1):ncol(tab))]
rownames(ft_plot) <- tab$SampleNo
ft_plot <- as.matrix(ft_plot)
ft_plot <- ft_plot[,-which(colSums(ft_plot)==0)]
```

```r
head(ft_plot)
```

```r
ids <- lapply( strsplit(colnames(ft_plot), split = '_'), function(x) x[1])
```

```r
colnames(ft_plot) <- paste(colnames(ft_plot), as.character(cf$CF_superclass[match(ids, as.character(cf$'cluster index'))]), sep = '_')
```

```r
chemclass <- as.character(cf$CF_superclass[match(ids, as.character(cf$'cluster index'))])
```

```r
ft_plot <- ft_plot[,-grep('no matches',colnames(ft_plot))]
chemclass <- chemclass[-grep('no matches', chemclass)]
```

```r
length(chemclass)
ncol(ft_plot)
```

order columns based on chemical classes

```r
ft_plot_ord <- ft_plot[,order(chemclass)]
```

```r
colnames(ft_plot_ord) <- lapply( strsplit(colnames(ft_plot_ord), split = '_'), function(x) x[1])
```

define colors for chemical classes

```r
#colchem <- hue_pal()(length(unique(chemclass)))
colchem <- brewer.pal(length(unique(chemclass)), 'Paired')
```

```r
chemclass <- chemclass[order(chemclass)]
```

```r
Colside_cols <- colchem[as.numeric(as.factor(chemclass))]
```

define color palette for heatmap

```r
#my_palette <- colorRampPalette(brewer.pal(11, "RdYlBu"))(n = 15)
my_palette <- viridis(15)
#my_palette <- colorRampPalette(c("blue", "white","red"))(n = 30) 
```

define colors for plant parts in the following order

```r
sort(unique(tab$TissueTreatment))
```

```r
#colpal <- hue_pal()(length(unique(tab$Tissue)))
colpal <- c('#4daf4a','#b2df8a','#a65628','#dfc27d','black','#999999')
#colpal <- c('#4daf4a','#b2df8a','#a65628','#dfc27d','darkgoldenrod','gold')
```

```r
Rowside_cols <- colpal[as.numeric(as.factor(tab$TissueTreatment))]
```

```r
pdf(file="2DModels_Heatmap_Superclasses_Ordered.pdf", width=80, height=20)
heatmap.2(ft_plot_ord, 
          cexCol = 0.5,
          cexRow = 0.5,scale="col", 
          margins = c(30, 50), 
          col = my_palette, 
          Rowv = FALSE, 
          Colv = FALSE, 
          tracecol = NA, 
          trace='none',
          density.info='none',
          ColSideColors=Colside_cols, 
          RowSideColors=Rowside_cols, 
          dendrogram = 'none', 
          keysize = 0.8,
          key.title = '',
          key.par = list(cex = 1.5),
          key = TRUE) #
legend("left", fill = colpal,
       legend = c("clipped leaves", "unclipped leaves", 
                  "clipped roots","unclipped roots",
                  "clipped stems","unclipped stems"),cex=2,horiz =FALSE)
legend("right", fill = colchem,
       legend = sort(unique(chemclass)),cex=1.5,horiz = F)
dev.off()
```
